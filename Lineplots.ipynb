### Importing Libraries

```python
#used to show figures in notebook
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as np
```

##### Creating a figure and an axes.

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x = np.linspace(0, 10, 1000)
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

### Adjusting the Plot: Line Colors and Styles

 Controlling the color of plot elements

```python
plt.plot(x,np.cos(x), color='blue')   # specify color by name     
plt.plot(x,np.cos(x + 1), color='g')      # short color code (rgbcmyk)    ----Red/Green/Blue/Cyan/Magenta/Yellow/blacK 
plt.plot(x,np.cos(x + 2), color='0.75')    # Grayscale between 0 and 1    
plt.plot(x,np.cos(x + 3), color='#FFDD44')   # Hex code (RRGGBB from 00 to FF)  
plt.plot(x,np.cos(x + 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x,np.cos(x + 5), color='chartreuse') # all HTML color names supported
```

various line styles

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

Controlling colors and styles with the shorthand syntax

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

###  Adjusting the Plot: Axes Limits

```python
plt.plot(x,np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
#Another method
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```
